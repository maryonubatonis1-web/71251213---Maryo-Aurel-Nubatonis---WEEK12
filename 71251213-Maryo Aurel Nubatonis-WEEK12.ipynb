```python
fname = input("Enter file name: ")
f = open(fname)

d = {}

for line in f:
    if line.startswith("From "):
        w = line.split()
        jam = w[5].split(":")[0]
        d[jam] = d.get(jam, 0) + 1

for k in sorted(d):
    print(k, d[k])
```
